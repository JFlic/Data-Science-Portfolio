```python
#Imports
import sqlite3
import pandas as pd
```

```python
#Importing Covid Deaths file with pandas into CD variable
CD = pd.read_csv("CovidDeaths.csv")

#creating connection variable
connection = sqlite3.connect('CovidDeaths.db')

#creating database
CD.to_sql('Covid_Deaths', connection, if_exists='replace')
```

```python
#closing connection
connection.close()
```

```python
#Importing Covid Vaccine file
CV = pd.read_csv("CovidVaccinations.csv")

connection = sqlite3.connect('CovidVaccinations.db')

CV.to_sql('Covid_Vaccinations', connection, if_exists='replace')
```

```python
connection.close()
```

```python
%%sql
Select location, date, total_cases, new_cases, total_deaths, (total_deaths/total_cases)*100 as DeathPercentage
From Covid_Deaths
Where continent is not null
LIMIT 50
```

```python
%%sql

Select SUM(new_cases) as total_cases, SUM(cast(new_deaths as int)) as total_deaths, SUM(cast(new_deaths as int))/SUM(New_Cases)*100 as DeathPercentage
From Covid_Deaths
where continent is not null
```

```python
#Shows total percentage of population that got Covid
%%sql

Select location, date, population, total_cases, (total_cases/population)*100 as Percent_Infected
From Covid_Deaths
Where location like '%states%'
LIMIT 50
```

```python
%%sql

Select location, population, MAX(total_cases) as HighestInfectionCount, Max((total_cases/population))*100 as Percent_Infected
From Covid_Deaths
Group by location, population
order by Percent_Infected desc
LIMIT 50
```

```python
%%sql

Select Location, Population,date, Max(total_cases) as HighestInfectionCount, Max((total_cases/population))*100 as PercentPopulationInfected
From Covid_Deaths
Group by Location, Population, date
order by PercentPopulationInfected desc
LIMIT 50
```

```python
#Showing Countries with Highest Death Count per Population

%%sql

Select location, MAX(cast(Total_deaths as int)) as TotalDeathCount
From Covid_Deaths
Where continent is not null
Group by location
order by TotalDeathCount desc
```

```python
#Showing Countries with Highest Death Count per continent
%%sql

Select continent, MAX(cast(Total_deaths as int)) as TotalDeathCount
From Covid_Deaths
Where continent is not null
Group by continent
order by TotalDeathCount desc
```

```python
#Total Global Numbers
%%sql

Select SUM(new_cases), SUM(cast(new_deaths as int)),
SUM(cast(new_deaths as int))/SUM(New_Cases)*100 as DeathPercentage
From Covid_Deaths
```
